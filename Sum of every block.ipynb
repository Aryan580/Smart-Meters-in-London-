```python
import pandas as pd
```

```python
DFS = []
```

```python
for i in range(0,112):
    fp = 'C:\\Users\\Aryan\\Desktop\\Data\\SML\\Daily Energy Used Dataset (Block-wise)\\'+str(i)+'.csv'
    df = pd.read_csv(fp)
    df = df.drop(['LCLid','day'],axis=1)
    a = df.sum()
    a = list(a)
    DFS.append(a)
```

```python
DFS
```

```python
DFS = pd.DataFrame(DFS)
```

```python
DFS = DFS.rename(index=str, columns={0:'energy_median',1:'energy_mean',2:'energy_max',3:'energy_count',4:'energy_std',5:'energy_sum',6:'energy_min'})
```
